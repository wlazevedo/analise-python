```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Carregar os datasets
df_casos = pd.read_excel('CasosSudeste2020.xlsx')
df_obitos = pd.read_excel('ObitosSudeste2020.xlsx')

# Ajuste do tamanho da interface grfica e adio das informaes dos dados coletados dos datasets
plt.figure(figsize=(16, 12))
plt.figtext(0.1, 0.02, "Sudeste: 27/03/2020 a 31/12/2020", fontsize=10, color='gray')
plt.figtext(0.1, 0.01, "Atualizao dos dados em 08/05/2024 s 15:31:27, com dados contidos nas Secretarias Estaduais de Sade.", fontsize=10, color='gray')

# Anlise de Casos Novos e bitos Novos por Semana Epidemiolgica
sns.scatterplot(x='Semana_Ano', y='Casos Novos', data=df_casos, color='blue', label='Casos Novos')
sns.scatterplot(x='Semana_Ano', y='Obitos_novos', data=df_obitos, color='red', label='bitos Novos')
plt.xticks(rotation=45)
plt.title('Casos Novos e bitos Novos por Semana Epidemiolgica na Regio Sudeste no ano de 2020 - MG, ES, RJ e SP')
plt.xlabel('Semana/Ano')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(True)  # Adiciona as grades ao grfico

# Adio da linha de regresso para Casos Novos
X_casos = np.arange(1, len(df_casos)+1).reshape(-1, 1)
y_casos = df_casos['Casos Novos']
model_casos = LinearRegression()
model_casos.fit(X_casos, y_casos)
plt.plot([1, len(df_casos)], [model_casos.predict([[1]]), model_casos.predict([[len(df_casos)]])], color='blue', label='Linha de Regresso - Casos Novos')

# Adio da linha de regresso para bitos Novos
X_obitos = np.arange(1, len(df_obitos)+1).reshape(-1, 1)
y_obitos = df_obitos['Obitos_novos']
model_obitos = LinearRegression()
model_obitos.fit(X_obitos, y_obitos)
plt.plot([1, len(df_obitos)], [model_obitos.predict([[1]]), model_obitos.predict([[len(df_obitos)]])], color='red', label='Linha de Regresso - bitos Novos')

# Carregar os datasets
df_casos = pd.read_excel('CasosSudeste2020.xlsx')
df_obitos = pd.read_excel('ObitosSudeste2020.xlsx')

# Criar o grfico de outliers para Casos Novos
plt.figure(figsize=(8, 6))
sns.boxplot(y='Casos Novos', data=df_casos, color='green', width=0.3)
plt.title('Grfico de Outliers - Casos Novos na Regio Sudeste no ano de 2020')
plt.ylabel('Casos Novos')
plt.grid(True)
plt.show()

# Criar o grfico de outliers para bitos Novos
plt.figure(figsize=(8, 6))
sns.boxplot(y='Obitos_novos', data=df_obitos, color='orange', width=0.3)
plt.title('Grfico de Outliers - bitos Novos na Regio Sudeste no ano de 2020')
plt.ylabel('bitos Novos')
plt.grid(True)
plt.show()
```
